# XGBoost Introduction 

### 1.Import dependencies 



```python
import pandas as pd 
import numpy as np 
from xgboost import XGBRegressor
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
PATH = "D:\\Coding_Stuff\\GitHub\\Machine_Learning_and_Optimization\\3.XGBoost Intro\\melb_data.csv"

data = pd.read_csv(PATH)
data.head()
```

```python
subset = data[['Rooms', 'Distance', 'Landsize', 'BuildingArea', 'YearBuilt', 'Price']]
subset.head()
```

```python
#split the data
from sklearn.model_selection import train_test_split

x = subset.iloc[: ,:-1]
y = subset.iloc[:, -1:]

X_train, X_test, y_train, y_test = train_test_split(
x, y, test_size=0.33, random_state=42)
```

```python
Xgboost = XGBRegressor(n_estimators = 100, learning_rate = 0.01)
Xgboost.fit(X_train,y_train)
```

```python
# Evaluation 

from sklearn.metrics import mean_absolute_error,r2_score

pred = Xgboost.predict(X_test)
print("Mean Absolute Score is: ", mean_absolute_error(y_test, pred))
print("R2 Score is: ", r2_score(y_test, pred))
```
